```python
import numpy as np
```

```python
train_features = np.load('data/train_features.npy')
train_labels = np.load('data/train_labels.npy')

test_features = np.load('data/test_features.npy')
test_labels = np.load('data/test_labels.npy')
```

(a)

```python
# We use linear regression, we have the closed-form solution:

class LinearRegression():
    
    def __init__(self):
        self.theta = 0

    def fit(self, X, y):
        
        self.X_new = []
        self.X_new = list(X.T)
        self.X_new.append(np.ones(X.T.shape[1:]))
        self.X_new = np.asarray(self.X_new)
        self.X_new = self.X_new.T
        
        self.theta = np.linalg.inv(self.X_new.T@self.X_new)@self.X_new.T@y
        
        self.MSE_score = np.sum(np.sum(np.power(self.X_new@self.theta-y, 2)))
        
    def predict(self, X):

        self.X_new = []
        self.X_new = list(X.T)
        self.X_new.append(np.ones(X.T.shape[1:]))
        self.X_new = np.asarray(self.X_new)
        self.X_new = self.X_new.T
                
        return self.X_new@self.theta
```

```python
X = train_features
y = train_labels

lg = LinearRegression()
lg.fit(X,y)

print('MSE value: \n', lg.MSE_score)
print('theta value: \n',lg.theta)
```

(b)

```python
X_test = test_features
y_test = test_labels

y_pred = lg.predict(X_test)
```

```python
def change_sign(x):
    if -x<0:
        return 0
    if -x >=0:
        return 1
        
np.sum(y_test-np.vectorize(change_sign)(y_pred)) / np.sum(y_test) *100
```

59% is not a much higher percentage than random prediction. Linear regression is not a good model.
