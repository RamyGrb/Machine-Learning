```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("data/train.csv")
```

```python
train.info()
```

```python
sns.barplot(x='Sex',y='Survived',data=train)
```

```python
sns.barplot(x='Pclass',y='Survived',hue='Sex',data=train)
```

```python
train.groupby('Sex').Survived.value_counts()
```

```python
sns.set_style('whitegrid')

for i in train.Pclass.unique():
    sns.distplot(train[train.Pclass == i].Age.dropna(),kde=True,hist=False)
    
```

```python
sns.countplot(x='Embarked',hue='Sex',data=train)
```

```python
sns.catplot(x='Embarked',hue='Sex',col='Survived',row='Pclass',data=train,kind='count')
```

```python
sns.set_style('darkgrid')
g = sns.FacetGrid(train,col='Survived',row='Sex')
g.map(sns.distplot,'Age',hist=False)
```

```python
sns.pairplot(train,hue='Survived')
```
