## Seaborn Overview

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('data/Pokemon.csv',encoding = "ISO-8859-1",index_col=0)
```

```python
df.head()
```

Quick plot

```python
#Default Scatterplot: the regression plot

# sns.lmplot(x=df.Attack, y=df.Defense)

sns.lmplot(
    x='Attack',y='Defense', # Dataframe columns to use
    data=df, # Data source
    fit_reg=False, # Remove the regression line
    hue='Stage') # Color by group of dataframe column

plt.ylim(0, 200) # Matplotlib editing
plt.xlim(0, 160) # Matplotlib editing
```

General Theme

```python
sns.set_style('whitegrid') # Darkgrid theme also available
stats_df = df.drop(['Total','Stage','Legendary'],axis=1)
sns.boxplot(data=stats_df)
```

Color

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

sns.swarmplot(x='Type 1', y='Attack', data=df, 
              palette=pkmn_type_colors)
```

Heatmap

```python
corr=stats_df.corr()
sns.heatmap(corr)
```

Histogram

```python
sns.distplot(df.Attack)
```

Barplot

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

Factor plot

```python
# Factor Plot
g = sns.factorplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)
```

Density

```python
sns.kdeplot(df.Attack, df.Defense)
```

Joint Distribution Plot

```python
sns.jointplot(x='Attack', y='Defense', data=df)
```
